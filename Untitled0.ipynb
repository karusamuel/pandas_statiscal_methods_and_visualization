```python
#convert to ksh and assign to a new column
ksh_rate = 130
df['Price in ksh'] = df['SalePrice'].apply(lambda x: x * ksh_rate)
df[['Price in ksh','SalePrice']].head() #for comparison
```

```python
#define function convert to also calculate to Kenyan rate
def convert(x):
  return x * ksh_rate #must return value

```

```python
#use function to calculate
df['Price in ksh'] = df['SalePrice'].apply(convert)
df[['Price in ksh','SalePrice']].head()
```
